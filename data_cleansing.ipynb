```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Housing.csv')
df.head()
```

```python
x = {
    'yes': 1,
    'no': 0,
}
x
```

```python
for index, row in df.iterrows():
    for index2, col in row.items():
        if isinstance(col, str):
            df.at[index, index2] = x.get(col, col)
```

```python
df = pd.get_dummies(df, columns=['furnishingstatus']).astype(int)
df.head()
```

```python
df.to_csv('Housing_Cleaned.csv', index=False, header=False)
```

```python
X = df.drop('price', axis=1)
y = df['price']
X.head()
```

```python
y.head()
```

```python
X_train = X[0:int(0.8*len(X))]
X_test = X[int(0.8*len(X)):]
y_train = y[0:int(0.8*len(y))]
y_test = y[int(0.8*len(y)):]
```

```python
print(len(X_train))
print(len(X_test))
print(len(y_train))
print(len(y_test))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
model = LinearRegression()
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)
r2_score(y_test, y_pred)
```
